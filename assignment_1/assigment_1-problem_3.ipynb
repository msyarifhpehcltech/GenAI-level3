# **3.	Problem Statement**

Write a Python program to draw (visualize) the architecture of a Neural Network used to classify fraudulent and non-fraudulent credit card transactions.

Assume the fraud detection dataset contains the following input features:
- TransactionAmount
- TransactionTime
- MerchantCategory
- CustomerAge
- AccountBalance
- NumberOfTransactionsToday
- Fraud (0 = Genuine, 1 = Fraud)



```python
!pip install tensorflow
```

```python
from tensorflow.keras.utils import plot_model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

model = Sequential([
    # TransactionAmount
    # TransactionTime
    # MerchantCategory
    # CustomerAge
    # AccountBalance
    # NumberOfTransactionsToday
    Input(shape=(6,)),

    # Hidden Layer 1, 8.
    Dense(8, activation='relu', name='Hidden_Layer_1'),

    # Hidden Layer 2, 4.
    Dense(4, activation='relu', name='Hidden_Layer_2'),

    # Output Layer, with sigmoid, Fraud (0/1)
    Dense(1, activation='sigmoid', name='Output_Layer')
])

plot_model(
    model,
    to_file='model.png',
    show_shapes=True,
    show_layer_names=True
)
```
